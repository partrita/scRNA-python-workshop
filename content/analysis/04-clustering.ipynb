# 클러스터링

## 소개

발현 프로파일의 유사성을 기반으로 세포를 그룹화하면 세포 유형과 상태를 식별하고 그룹 간의 차이를 추론할 수 있습니다. 이는 __클러스터링__ 또는 __커뮤니티 탐지__를 통해 수행됩니다. 

__비지도 클러스터링__은 다양한 응용 분야에서 유용하며 기계 학습에서 널리 연구되었습니다. 가장 인기 있는 접근 방식 중 일부는 __계층적 클러스터링__과 __k-평균 클러스터링__입니다. 이러한 방법은 세포 간의 __거리 메트릭__을 계산하고(종종 PCA, tSNE 또는 UMAP과 같은 저차원 표현을 기반으로 함) 이러한 거리를 기반으로 세포를 반복적으로 그룹화합니다.

__커뮤니티 탐지__ 방법(그래프 기반 클러스터링이라고도 함)은 __이웃 그래프__를 분할합니다. 이웃 그래프는 각 세포를 노드로 취급하며, 각 노드를 _k_개의 가장 가까운 이웃(위에서 설명한 유사한 거리 메트릭 기준)에 연결하는 엣지가 있습니다. 그런 다음 그래프는 이러한 연결성을 기반으로 모듈(즉, 세포 그룹/클러스터)로 분할됩니다. 이러한 방법은 일반적으로 유사한 효과를 가진 다른 클러스터링 방법보다 빠릅니다.  

거의 모든 클러스터링 또는 커뮤니티 탐지 방법이 공통적으로 가지고 있는 것은 일종의 __해상도 매개변수__입니다. 이 매개변수는 추론된 클러스터가 얼마나 세분화되거나 거친지를 제어합니다. 이 매개변수는 결과에 큰 영향을 미칠 수 있습니다! 아래에서 이에 대해 자세히 살펴보겠습니다.

여기서는 __k-평균 클러스터링__과 그래프 기반 __루뱅 클러스터링__ 방법을 살펴보겠습니다.
